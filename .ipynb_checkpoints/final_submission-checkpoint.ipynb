```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.simplefilter(action='ignore')


from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, VarianceThreshold, f_regression
from sklearn.linear_model import ElasticNet

pd.options.display.max_columns = 82
pd.options.display.max_rows = 1000

%matplotlib inline
```

```python
# train data
data = pd.read_csv('./datasets/train.csv')

# test data
test_data = pd.read_csv('./datasets/test.csv')
```

# Step 2: data cleaning

## 2.1 : drop columns with null greater than 1000

```python
drop_column = [column for column in data.columns if data[column].isnull().sum() >= 1000]

data = data.drop(drop_column , axis = 1)
```

## 2.2.1:  clean null values (training)

```python
# fill na for Mas Vnr Type and Mas Vnr Area

#check null individually
data.loc[data['Mas Vnr Type'].isnull()].shape
data.loc[data['Mas Vnr Area'].isnull()].shape

#check null in pair
data.loc[(data['Mas Vnr Type'].isnull()) & (data['Mas Vnr Area'].isnull())].shape

# fill null with 'None' and 0
data['Mas Vnr Type'].fillna('None',inplace = True)
data['Mas Vnr Area'].fillna(0,inplace = True)

# fill na for BsmtFin SF 1, BsmtFin SF 2, Bsmt Unf SF, Total Bsmt SF, Bsmt Full Bath, Bsmt Half Bath
# drop na Garage Cars, Garage Area

# fill na
data['BsmtFin SF 1'].fillna(0, inplace = True)
data['BsmtFin SF 2'].fillna(0, inplace = True)
data['Bsmt Unf SF'].fillna(0, inplace = True)
data['Total Bsmt SF'].fillna(0, inplace = True)
data['Bsmt Full Bath'].fillna(0, inplace = True)
data['Bsmt Half Bath'].fillna(0, inplace = True)

# drop na
data.dropna(subset = ['Garage Cars','Garage Area'], inplace = True)


# fill 'NA' for Bsmt Qual,Bsmt Cond, Bsmt Exposure 

# check null individually
data.loc[(data['Bsmt Qual'].isnull())].shape
data.loc[(data['Bsmt Cond'].isnull())].shape
data.loc[(data['Bsmt Exposure'].isnull())].shape   # 3 extra null


# check null in pair
data.loc[(data['Bsmt Qual'].isnull()) & 
         (data['Bsmt Cond'].isnull()) & 
         (data['Bsmt Exposure'].isnull())].shape

# check value counts in each column
data['Bsmt Qual'].value_counts()
data['Bsmt Cond'].value_counts()
data['Bsmt Exposure'].value_counts()


# check 3 extra Bsmt Exposure null with given Qual Cond
data.loc[(data['Bsmt Qual'] == 'Gd') & 
         (data['Bsmt Cond'] == 'TA') & 
         (data['Bsmt Exposure'].isnull()), 'Bsmt Exposure'] = 'No'

# replace Nan with 'NA' for Bsmt Qual, Bsmt Cond
data['Bsmt Qual'].fillna('NA',inplace = True)
data['Bsmt Cond'].fillna('NA',inplace = True)
data['Bsmt Exposure'].fillna('NA',inplace = True)


# fill 'NA' for Garage Type, Garage Finish, Garage Qual, Garage Cond
# fill np.nan for Garage Yr Blt

# check null individually
data.loc[(data['Garage Type'].isnull())].shape
data.loc[(data['Garage Yr Blt'].isnull())].shape
data.loc[(data['Garage Finish'].isnull())].shape
data.loc[(data['Garage Cond'].isnull())].shape


# check null in pair
data.loc[(data['Garage Type'].isnull()) & 
         (data['Garage Yr Blt'].isnull()) &
         (data['Garage Finish'].isnull()) & 
         (data['Garage Cond'].isnull())]


data['Garage Type'].fillna('NA', inplace = True)
data['Garage Finish'].fillna('NA', inplace = True)
data['Garage Qual'].fillna('NA', inplace = True)
data['Garage Cond'].fillna('NA', inplace = True)
data['Garage Yr Blt'].fillna(0, inplace = True)


# fill 'NA' for BsmtFin Type 1, BsmtFin Type 2

# determine what to fill in special null for bsmtfin type 2
data.loc[(data['BsmtFin Type 1'].notnull()) & (data['BsmtFin Type 2'].isnull())]
data.groupby('BsmtFin Type 2')['BsmtFin SF 2'].mean()

# bsmtfin type 2 fill in ALQ when type 1 is not null:
data.loc[(data['BsmtFin Type 1'].notnull()) & (data['BsmtFin Type 2'].isnull()), 'BsmtFin Type 2'] = 'ALQ'

# fill 'NA' for BsmtFin Type 1, BsmtFin Type 2
data['BsmtFin Type 1'].fillna('NA',inplace = True)
data['BsmtFin Type 2'].fillna('NA',inplace = True)

# set Lot Frontage null to 0
data['Lot Frontage'].fillna(0, inplace = True)

# drop outlier: living area > 4000
data.drop(labels = data[data['Gr Liv Area']>4000].index,inplace = True)
```

## Model 1: use all numeric value

```python
columns = [column for column in data._get_numeric_data().columns]
```

### Model 1 Linear Regression

```python
# set X and y
X = data.loc[: , columns].drop('SalePrice', axis = 1)
y = data['SalePrice']

# K-fold cross validation
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)

# fit Linear Regression
lm = LinearRegression()
lm.fit(X_train, y_train)

# cross validation for R2 training dataset
R2_train_lm = cross_val_score(lm, X_train, y_train, cv = 5).mean()

# cross validation for R2 testing dataset
R2_test_lm = cross_val_score(lm, X_test, y_test, cv = 5).mean()

# predicted y_test
y_preds = lm.predict(X_test)


# RMSE
RMSE = np.sqrt(mean_squared_error(y_test, y_preds))

# Coef
coef = lm.coef_



print (f'RMSE: {RMSE}')
print (f'R2 training: {R2_train_lm}')
print (f'R2 test: {R2_test_lm}')
```

```python
# plot actual vs pred sale price
plt.figure(figsize = (8,8))
plt.hist(y_test, bins = 50, color = 'black', alpha = 1, label ='actual')
plt.hist(y_preds, bins = 50, color = 'red', alpha = 0.7, label = 'prediction')
plt.title('Model 1: Linear Regression Accuracy', size = 20)
plt.legend()
;
```

```python
# equal variance
plt.scatter(y_preds, y_test - y_preds)
plt.axhline(0,color = 'r', alpha = 0.9)
plt.title('Model 1: Linear Regression Residual plot', size = 20)
;
```

```python
# normality check
plt.hist(y_test - y_preds)
plt.axvline(0, color = 'r', alpha = 0.9)
plt.title('Distribution', size = 20)
;
```

### Model 1 Lasso

```python
# Standardize dataset
ss = StandardScaler()
X_train_lasso = ss.fit_transform(X_train)
X_test_lasso = ss.transform(X_test)

# Lasso model
model_lasso = Lasso(alpha = 281)
model_lasso.fit(X_train_lasso, y_train)

# R2 for training
R2_train_lasso = cross_val_score(model_lasso, X_train_lasso, y_train, cv=5).mean()

# R2 for testing
R2_test_lasso = cross_val_score(model_lasso, X_test_lasso, y_test, cv=5).mean()

# predict y
y_preds_lasso = model_lasso.predict(X_test_lasso)

# RMSE
RMSE_lasso = np.sqrt(mean_squared_error(y_test,y_preds_lasso))

print (f'RMSE: {RMSE_lasso}')
print (f'R2 training: {R2_train_lasso}')
print (f'R2 test: {R2_test_lasso}')
```

```python
# plot actual vs pred sale price
plt.figure(figsize = (8,8))
plt.hist(y_test, bins = 50, color = 'black', alpha = 1, label ='actual')
plt.hist(y_preds_lasso, bins = 50, color = 'red', alpha = 0.7, label = 'prediction')
plt.title('Model 1: Lasso Accuracy', size = 20)
plt.legend()
;
```

```python
# equal variance
plt.scatter(y_preds_lasso, y_test - y_preds_lasso)
plt.axhline(0,color = 'r', alpha = 0.9)
plt.title('Model 1: Lasso Regression Residual plot', size = 20)
;
```

## Model 2: high correlated num values + square & interaction

```python
# Identify numeric columns with high correlation
corr_col = {}
for i in range(len(data.corr()['SalePrice'])):
    if abs(data.corr()['SalePrice'].values[i]) > 0.5:
         corr_col[data.corr()['SalePrice'].index[i]] = data.corr()['SalePrice'].values[i]
del corr_col['SalePrice']

# add columns into a list
corr_list = [col for col in corr_col.keys()]
train_df = data[corr_list]
```

```python
# heatmap for num columns with high correlation
plt.figure(figsize = (10,10))
sns.heatmap(train_df.corr(), vmin = -1, vmax = 1,cmap = 'coolwarm', annot = True)


# Code adapt from class notes
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t); # update the ylim(bottom, top) values

;
```

```python
# add square to column
list_of_columns = list(train_df.columns)
for i in list_of_columns:
    train_df[str(i + '^2')] = train_df[i] * train_df[i]
```

```python
# add interaction based on correlation above 0.8
train_df['TotRms AbvGrd * Gr Liv Area'] = train_df['TotRms AbvGrd'] * train_df['Gr Liv Area']
train_df['Garage Area * Garage Cars'] = train_df['Garage Area'] * train_df['Garage Cars']
```

```python
ss = StandardScaler()

# set X and y
X = train_df
y = data['SalePrice']

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 ,random_state = 42)


# Scale X
X_train_ridge = ss.fit_transform(X_train)
X_test_ridge = ss.transform(X_test)

ridge_alphas = np.logspace(0, 3, 100)


ridge_cv = RidgeCV(
    alphas = ridge_alphas ,
    scoring = 'r2',
    cv = 5
)


# Fit model using best ridge alpha!
ridge_cv.fit(X_train_ridge, y_train)

# find best alpha
ridge_best_alpha = ridge_cv.alpha_

# R2 for training
R2_train_ridge_cv =ridge_cv.score(X_train_ridge, y_train)

# R2 for testing
R2_test_ridge_cv = ridge_cv.score(X_test_ridge, y_test)

# predict y
y_preds_ridge_cv = ridge_cv.predict(X_test_ridge)

# RMSE
RMSE_ridge_cv = np.sqrt(mean_squared_error(y_test,y_preds_ridge_cv))

print (f'RMSE: {RMSE_ridge_cv}')
print (f'R2 training: {R2_train_ridge_cv}')
print (f'R2 test: {R2_test_ridge_cv}')
```
